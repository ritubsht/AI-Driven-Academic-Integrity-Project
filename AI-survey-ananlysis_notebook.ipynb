```python
import matplotlib.pyplot as plt

def plot_likert_counts(df, col, title):
    counts = df[col].value_counts().sort_index()
    plt.figure(figsize=(5,4))
    counts.plot(kind="bar", color='g')
    plt.xlabel("Response (1–5)")
    plt.ylabel("Number of students")
    plt.title(title)
    plt.tight_layout()
    plt.show()

plot_likert_counts(df, "q1_familiar_integrity", "Q1: Familiarity with academic integrity")
plot_likert_counts(df, "q2_awareness_policies", "Q2: Awareness of policies")
plot_likert_counts(df, "q7_ai_equals_plagiarism", "Q7: AI use as serious as plagiarism")
plot_likert_counts(df, "q9_confidence_ai_integrity", "Q9: Confidence about AI & integrity")
```

```python
#The pre‑intervention survey of 91 students showed that 
#participants had a moderate understanding of academic integrity overall (Q1: M = 3.02, SD = 1.26 on a 1–5 scale). 
#Their awareness of specific institutional rules on plagiarism and AI use was slightly lower (Q2: M = 2.93, SD = 1.18),
#indicating that, although they recognise the general concept of integrity, many are not fully familiar with the detailed policies. 
#Students tended to agree that using AI tools without acknowledgment is comparable in seriousness to conventional plagiarism (Q7: M = 3.93, SD = 0.95), 
#suggesting that undisclosed AI assistance is widely perceived as a form of misconduct. 
#They also reported relatively strong confidence in their ability to distinguish appropriate from inappropriate AI use in academic work (Q9: M = 3.68, SD = 0.95), 
#although the variability in responses shows that a number of students still feel only moderately sure about where to draw the line.

#Overall, these findings imply that the student group enters the intervention with a basic grounding in academic integrity and a generally critical view of hidden AI use,
#but with incomplete knowledge of formal regulations and some uncertainty about how to apply ethical principles in real situations.
#This underscores the importance of targeted awareness activities within the PAIEL framework and of clear communication of institutional expectations regarding AI in academic tasks.
```

```python
def plot_category_percent(df, col, title):
    counts = df[col].value_counts(dropna=False)
    perc = (counts / len(df) * 100).round(1)
    plt.figure(figsize=(6,4))
    perc.plot(kind="bar", color='b')
    plt.ylabel("Percentage of students (%)")
    plt.title(title)
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()
    
plot_category_percent(df, "q3_aware_consequences", "Q3: Awareness of consequences")
plot_category_percent(df, "q4_used_ai", "Q4: Have you ever used AI tools for academic work?")
plot_category_percent(df, "q5_ai_acceptable", "Q5: Is AI-generated content acceptable?")
plot_category_percent(df, "q8_action_ai_integrity", "Q8: What would you do if a friend plagiarised?")
```

```python

# Split each row on ';' and store all reasons in a single list
all_reasons = []
for cell in df['q6_ai_ethics']:
    for r in str(cell).split(";"):
        r = r.strip()
        if r:
            all_reasons.append(r)

# Now count using value_counts
reason_counts = pd.Series(all_reasons).value_counts()

# Make a small summary table
q6_summary = pd.DataFrame({
    "count": reason_counts,
    "percent": (reason_counts / len(df) * 100).round(1)
})

print(q6_summary)

plt.figure(figsize=(5,4))
reason_counts.plot(kind="bar", color='g')
plt.xlabel("Response (1–5)")
plt.ylabel("Number of students")
plt.title('Reasons for using AI')
plt.tight_layout()
plt.show()


```

```python
t=df['q10_students_suggestions'].dropna() #remove the blank rows
t
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
import re
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords

clean_text =[]
for i in t:
  clean_text.append(i.lower())
  print(clean_text)
```

```python
tokenized_text = []
for i in clean_text:
  tokenized_text.append(word_tokenize(i))
```

```python
stop_words = set(stopwords.words('english'))
new_text = []
for i in tokenized_text :
  for x in i:
    if x not in stop_words:
      new_text.append(x)
```

```python
lem = WordNetLemmatizer()
for i in new_text:
  print(i,"<---------->",lem.lemmatize(i))
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
text_to_plot = " ".join(new_text)
```

```python
wordcloud = WordCloud(
    width = 800,
    height = 400,
    background_color = 'white',
    colormap ='viridis',
    prefer_horizontal = 1.0,
    collocations = False
).generate(text_to_plot)

plt.figure(figsize=(10,10))
plt.imshow(wordcloud)
plt.axis('off')
plt.title("Word Cloud")
plt.show()
```
