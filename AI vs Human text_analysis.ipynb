```python
import seaborn as sns
import matplotlib.pyplot as plt

confusion = pd.crosstab(df['author_type'], df['model_source'])
sns.heatmap(confusion, annot=True, fmt='d', cmap='Blues')
plt.title('AI Detector Error Patterns (Real-world Failures)')
plt.ylabel('True Author')
plt.xlabel('Detector Prediction')
plt.show()
```

```python
plt.figure(figsize=(6,4))
df['author_type'].value_counts().plot(kind='bar', color=['green','red'])
plt.title('Dataset Balance: Human vs AI')
plt.ylabel('Count')
plt.show()
```

```python
df['text_len'] = df['content_text'].str.len()
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='author_type', y='text_len')
plt.title('AI vs Human: Text Length')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# YOUR top features + target
features_to_test = ['prompt_complexity_score', 'perplexity_score', 'burstiness_index', 
                   'syntactic_variability', 'semantic_coherence_score', 
                   'lexical_diversity_ratio', 'readability_grade_level', 
                   'generation_confidence_score']

# Calculate MEANS for Human vs AI
heatmap_data = df.groupby('author_type')[features_to_test].mean().T

# Plot heatmap (RED = AI HIGH, BLUE = HUMAN HIGH)
plt.figure(figsize=(10,8))
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', center=0, 
            fmt='.3f', cbar_kws={'label': 'Mean Score'})
plt.title('FEATURE POWER: Red=AI-High, Blue=Human-High')
plt.ylabel('PAIEL Features')
plt.xlabel('Author Type')
plt.tight_layout()
plt.show()
```

```python
top3 = ['burstiness_index', 'perplexity_score', 'generation_confidence_score']
fig, axes = plt.subplots(1,3, figsize=(15,4))

for i, feature in enumerate(top3):
    sns.boxplot(data=df, x='author_type', y=feature, ax=axes[i])
    axes[i].set_title(feature)

plt.tight_layout()
plt.show()
```
